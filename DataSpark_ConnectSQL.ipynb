```python
myconn=pymysql.connect(host='host',user='user',passwd='********',database='dataspark_db')
```

```python
#creating the customer table in sql:

create_table_sql = """
CREATE TABLE customer (
    CustomerKey INT NOT NULL PRIMARY KEY,
    Gender TEXT,
    Name TEXT,
    City TEXT,
    State_Code TEXT,
    State TEXT,
    Zip_Code TEXT,
    Country TEXT,
    Continent TEXT,
    Birthday DATE
);
"""
print(create_table_sql)
myconn.cursor().execute(create_table_sql)
myconn.commit()
```

```python
#creating the product table in sql:

create_table_sql = """
CREATE TABLE product (
    ProductKey INT NOT NULL PRIMARY KEY,        
    Product_Name  TEXT,     
    Brand  TEXT,           
    Color TEXT,             
    Unit_Cost_USD   DECIMAL(7,2), 
    Unit_Price_USD  DECIMAL(7,2),  
    SubcategoryKey   INT , 
    Subcategory   TEXT,     
    CategoryKey    INT,     
    Category       TEXT    
    );
"""
print(create_table_sql)
myconn.cursor().execute(create_table_sql)
myconn.commit()
```

```python
#creating the exchange table in sql:

create_table_sql = """
CREATE TABLE exchange (
    Date   DATE,
    Currency  TEXT,
    Exchange  DECIMAL(7,2)
);
"""
print(create_table_sql)
myconn.cursor().execute(create_table_sql)
myconn.commit()
```

```python
#creating the stores table in sql:

create_table_sql = """
CREATE TABLE store(
    StoreKey           INT NOT NULL PRIMARY KEY,
    Country           TEXT,
    State             TEXT,
    Square_Meters     DECIMAL(7,2),
    Open_Date         DATE  
    );
"""
print(create_table_sql)
myconn.cursor().execute(create_table_sql)
myconn.commit()
```

```python
#creating the sales table in sql:

create_table_sql = """
CREATE TABLE sales(
    Order_Number       INT NOT NULL ,
    Line_Item          INT  NOT NULL,
    Order_Date        DATE,
    CustomerKey        INT,
    StoreKey           INT,
    ProductKey         INT,
    Quantity           INT,
    Currency_Code     TEXT,
    Exchange         DECIMAL(7,2),
    PRIMARY KEY (Order_Number, Line_Item) 
    
    );
"""

print(create_table_sql)
myconn.cursor().execute(create_table_sql)
myconn.commit()
```

```python
# Define the mapping between CSV column names and SQL table column names
column_mapping = {
    'State Code': 'State_Code',
    'Zip Code': 'Zip_Code'
    
}

# Database connection string :
database_url='mysql+pymysql://user:pwd@host/dataspark_db'

# Create a SQLAlchemy engine
engine =create_engine(database_url)

# Table name where data will be inserted
table_name = 'customer'

# Rename DataFrame columns based on the mapping
df_customer.rename(columns=column_mapping, inplace=True)


# Insert the DataFrame into the SQL table
df_customer.to_sql(table_name, engine, if_exists='append', index=False)

print(f"Data from dataframe inserted into table '{table_name}'.")
```
